# (Data frame merge)


```python
#q1
df1 = pd.DataFrame({'A': [1, 2, 3, 4], 'B': ['a', 'b', 'c', 'd']})
df1
```

```python
df2 = pd.DataFrame({'A': [1, 2, 5, 6], 'C': [10, 20, 30, 40]})
df2
```

```python
#q2
pd.merge(df1,df2)
```

# Regression plots)

```python
#q1
sns.lmplot(x='carat', y='price', data=diamonds, ci=95)
```

```python
#q2
# Scatter plot
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.3, label='Data')
```

```python
# Linear regression line
linear_coeffs = np.polyfit(diamonds['carat'], diamonds['price'], deg=1)
linear_trend = np.poly1d(linear_coeffs)
plt.plot(diamonds['carat'], linear_trend(diamonds['carat']), label='Linear')
```

```python
# Cubic trend
cubic_coeffs = np.polyfit(diamonds['carat'], diamonds['price'], deg=3)
cubic_trend = np.poly1d(cubic_coeffs)
plt.plot(diamonds['carat'], cubic_trend(diamonds['carat']), label='Cubic')
```

```python
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Custom Linear Regression Plot')
plt.legend()
plt.show()
```

```python
linear_coeffs = np.polyfit(diamonds['carat'], diamonds['price'], deg=1)
linear_trend = np.poly1d(linear_coeffs)
plt.plot(diamonds['carat'], linear_trend(diamonds['carat']), label='Linear')

# Quadratic trend
quadratic_coeffs = np.polyfit(diamonds['carat'], diamonds['price'], deg=2)
quadratic_trend = np.poly1d(quadratic_coeffs)
plt.plot(diamonds['carat'], quadratic_trend(diamonds['carat']), label='Quadratic')

# Cubic trend
cubic_coeffs = np.polyfit(diamonds['carat'], diamonds['price'], deg=3)
cubic_trend = np.poly1d(cubic_coeffs)
plt.plot(diamonds['carat'], cubic_trend(diamonds['carat']), label='Cubic')

plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Custom Linear Regression Plot')
plt.legend()
plt.show()
```
